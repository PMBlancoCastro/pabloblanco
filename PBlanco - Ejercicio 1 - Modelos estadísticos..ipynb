## Ejercicio 1.3

Generamos el proceso Autoregresivo 

```r
w = rnorm(150,0,1) # 50 extra to avoid startup problems
x = filter(w, filter=c(1,-.9), method="recursive")[-(1:50)] # remove first 50
plot.ts(x, main="autoregression")
```

Generamos el proceso de media móvil

```r
v = filter(x, rep(1/4, 4), sides = 2)
plot.ts(x)
lines(v, col=2)
```

Para el caso B

```r
x=cos(2*pi*1:100/4)
b=filter(x, rep(1/4,4), sides=1)
plot.ts(x, main='Coseno')
lines (b, col=2)
```

Para el caso C

```r
w=rnorm(100,0,1)
x=cos(2*pi*1:100/4) + w
a=filter(x, rep(1/4,4), sides=2)
plot.ts(x, main='Coseno')
lines (a, col=2)
```

Podemos apreciar que agregando el componente de media móvil y el paseo aleatorio a la serie de coseno hace que sea incluso dificultoso reconocerla, cambiando totalmente el gráfico de la serie.
